```python
# Celda 1: Configuraci√≥n inicial
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

plt.rcParams['figure.figsize'] = [12, 4]
plt.rcParams['font.size'] = 12# Celda 3: Visualizaci√≥n de las primeras 5 iteraciones
print("üîç CONSTRUCCI√ìN DEL CONJUNTO DE CANTOR")
print("=" * 50)

total_removed, final_length = plot_cantor(5)

print(f"\nüìä RESUMEN MATEM√ÅTICO:")
print(f"‚Ä¢ Longitud total removida: {total_removed:.10f}")
print(f"‚Ä¢ Longitud que queda: {final_length:.10f}")
print(f"‚Ä¢ L√≠mite cuando n ‚Üí ‚àû: {(1/3) / (1 - 2/3):.10f}")
```

```python
# Celda 2: Funci√≥n para visualizar el conjunto de Cantor
def plot_cantor(iterations=5):
    fig, ax = plt.subplots(1, 1, figsize=(15, 6))
    
    # Color y altura para cada iteraci√≥n
    colors = plt.cm.viridis(np.linspace(0, 1, iterations))
    height = 0.8 / iterations
    
    total_removed = 0
    remaining_length = 1
    
    for i in range(iterations):
        # Calcular segmentos actuales
        segments = get_cantor_segments(i)
        y_pos = 1 - (i * height) - height/2
        
        # Dibujar segmentos que quedan
        for start, end in segments:
            length = end - start
            rect = patches.Rectangle((start, y_pos), length, height*0.8,
                                   facecolor=colors[i], alpha=0.7,
                                   edgecolor='black', linewidth=1)
            ax.add_patch(rect)
        
        # Calcular longitud removida en esta iteraci√≥n
        if i == 0:
            removed_this_step = 1/3
        else:
            removed_this_step = (2**(i-1)) * (1/(3**i))
        
        total_removed += removed_this_step
        remaining_length -= removed_this_step
        
        # Anotaciones
        ax.text(1.05, y_pos + height/2, 
                f'Iter {i}: Queda {remaining_length:.4f}', 
                va='center')
    
    ax.set_xlim(-0.1, 1.3)
    ax.set_ylim(0, 1)
    ax.set_xlabel('Segmento [0,1]')
    ax.set_ylabel('Iteraciones')
    ax.set_title('Construcci√≥n del Conjunto de Cantor - Proceso Paso a Paso')
    ax.grid(True, alpha=0.3)
    
    # Leyenda de longitud
    ax.text(0.5, -0.15, f'Longitud total removida: {total_removed:.6f}\nLongitud final: {remaining_length:.6f}', 
            ha='center', transform=ax.transAxes, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))
    
    plt.tight_layout()
    plt.show()
    
    return total_removed, remaining_length

def get_cantor_segments(iteration):
    if iteration == 0:
        return [(0, 1)]
    
    previous_segments = get_cantor_segments(iteration - 1)
    new_segments = []
    
    for start, end in previous_segments:
        length = end - start
        # Dividir cada segmento en tercios y mantener solo el primero y √∫ltimo
        new_segments.append((start, start + length/3))
        new_segments.append((start + 2*length/3, end))
    
    return new_segments
```

```python
# Celda 3: Visualizaci√≥n de las primeras 5 iteraciones
print("üîç CONSTRUCCI√ìN DEL CONJUNTO DE CANTOR")
print("=" * 50)

total_removed, final_length = plot_cantor(5)

print(f"\nüìä RESUMEN MATEM√ÅTICO:")
print(f"‚Ä¢ Longitud total removida: {total_removed:.10f}")
print(f"‚Ä¢ Longitud que queda: {final_length:.10f}")
print(f"‚Ä¢ L√≠mite cuando n ‚Üí ‚àû: {(1/3) / (1 - 2/3):.10f}")
```

```python
# Celda 4: Animaci√≥n del proceso (versi√≥n est√°tica m√∫ltiple)
def plot_multiple_iterations():
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    
    for i, ax in enumerate(axes):
        if i < 6:  # Mostrar iteraciones 0-5
            segments = get_cantor_segments(i)
            height = 0.6
            
            for start, end in segments:
                length = end - start
                rect = patches.Rectangle((start, 0.2), length, height,
                                       facecolor='skyblue', alpha=0.8,
                                       edgecolor='navy', linewidth=2)
                ax.add_patch(rect)
            
            remaining_length = sum(end - start for start, end in segments)
            ax.set_xlim(-0.1, 1.1)
            ax.set_ylim(0, 1)
            ax.set_title(f'Iteraci√≥n {i}\nLongitud = {remaining_length:.6f}')
            ax.set_aspect('equal')
            ax.grid(True, alpha=0.3)
        else:
            ax.axis('off')
    
    plt.suptitle('Evoluci√≥n del Conjunto de Cantor - De 1 a ‚âà0', fontsize=16, y=0.95)
    plt.tight_layout()
    plt.show()

plot_multiple_iterations()
```

```python
# Celda 5: C√°lculo detallado de la progresi√≥n geom√©trica
print("üßÆ C√ÅLCULO DETALLADO DE LA LONGITUD")
print("=" * 40)

longitud_inicial = 1
longitud_removida_por_paso = []
longitud_acumulada = []

for n in range(10):
    if n == 0:
        removida = 1/3
    else:
        removida = (2**(n-1)) * (1/(3**n))
    
    longitud_removida_por_paso.append(removida)
    longitud_acumulada.append(sum(longitud_removida_por_paso))
    
    print(f"Paso {n}: Removido = {removida:.8f} | Acumulado = {longitud_acumulada[-1]:.8f}")

# Gr√°fico de la progresi√≥n
plt.figure(figsize=(10, 6))
plt.plot(range(10), longitud_acumulada, 'o-', linewidth=2, markersize=8, label='Longitud removida acumulada')
plt.axhline(y=1, color='red', linestyle='--', linewidth=2, label='Longitud total (1)')
plt.xlabel('Iteraci√≥n')
plt.ylabel('Longitud')
plt.title('Progresi√≥n de la Longitud Removida en el Conjunto de Cantor')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nüí° Cuando n ‚Üí ‚àû, la serie converge a: {sum(longitud_removida_por_paso):.10f}")
```

```python
# Celda 6: Analog√≠a con las manzanas - Visualizaci√≥n conceptual
def analogia_manzanas():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # Lado izquierdo: Analog√≠a de las manzanas
    ax1.set_xlim(0, 10)
    ax1.set_ylim(0, 10)
    
    # Manzana completa inicial
    circle1 = plt.Circle((2, 5), 1.5, color='red', alpha=0.7)
    ax1.add_patch(circle1)
    ax1.text(2, 2, 'Manzana completa\nPeso = 1 kg', ha='center')
    
    # Flecha de proceso
    ax1.arrow(4, 5, 2, 0, head_width=0.3, head_length=0.2, fc='green', ec='green')
    
    # Manzana despu√©s de varias "peladas"
    circle2 = plt.Circle((8, 5), 0.3, color='red', alpha=0.7)
    ax1.add_patch(circle2)
    ax1.text(8, 2, 'Infinitas migajas\nPeso total ‚âà 0 kg', ha='center')
    
    ax1.set_title(' Analog√≠a con las Manzanas\n(Infinitas part√≠culas, peso cero)')
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    # Lado derecho: Matem√°tica del conjunto de Cantor
    segments = get_cantor_segments(4)
    for start, end in segments:
        length = end - start
        rect = patches.Rectangle((start*8 + 1, 3), length*8, 4,
                               facecolor='lightblue', alpha=0.8,
                               edgecolor='darkblue', linewidth=1)
        ax2.add_patch(rect)
    
    ax2.set_xlim(0, 10)
    ax2.set_ylim(0, 10)
    ax2.set_title(' Conjunto de Cantor\n(Infinitos puntos, longitud cero)')
    ax2.text(5, 1, f'{len(segments)} segmentos | Longitud total = {sum(end-start for start,end in segments):.6f}', 
             ha='center', bbox=dict(boxstyle="round", facecolor="yellow"))
    ax2.set_aspect('equal')
    ax2.axis('off')
    
    plt.suptitle('COMPARACI√ìN CONCEPTUAL: El Misterio de Medida Cero', fontsize=16, y=0.95)
    plt.tight_layout()
    plt.show()

analogia_manzanas()
```
